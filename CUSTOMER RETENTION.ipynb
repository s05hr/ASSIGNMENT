# DATA VISUALIZATION

```python
#countplot of gender column using seaborn library
plt.figure(figsize=(5,3))
a=sns.countplot(x='gender',data=df)
plt.title('Sex Distribution')
plt.xlabel('Gender')
plt.xticks([0, 1], ['Male', 'Female'])
plt.show()
df['gender'].value_counts()  #value counts of gender column
```

From the given data of 269:180 are females and 89 are males who shopped online.
This bar chart shows more females shop online than males.

```python
#countplot of age column using seaborn library
plt.figure(figsize=(20,5))
sns.countplot(x='age',data=df)
plt.title('Age Distribution')
plt.xlabel('Age group')
plt.xticks([0,1,2,3,4],['<20years','21-30years','31-40years','41-50years','51years and above'])
plt.show()
df['age'].value_counts() #value counts of age column
```

Age group distributionof people who shop online:
-20 people in the age less than 20 years(1)
-79 people in the age group of 21-30years(2)
-81 people in the age group of 31-40years(3)
-70 people in the age group of 41-50years(4)
-19 people in the age of 51years and above(5)

This shows maximum people of age group of 31-40 years shops online.
Data shows which age group shops online more:
31-40yrs > 21-30yrs > 41-50yrs > less than 20yrs > 51yrsandabove

```python
#age vs gender
plt.figure(figsize=(15,8))
a=sns.countplot(x='age',hue='gender',data=df)
plt.title('Distribution')
plt.xticks([0,1,2,3,4],['<20years','21-30years','31-40years','41-50years','51years and above'])
plt.show()
for i in a.containers:
    print(a.bar_label(i,))
```
