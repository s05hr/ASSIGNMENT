1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
import selenium     #import respective libraries for web scraping using selenium
import pandas as pd 
from selenium import webdriver 
import warnings    
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException 
from selenium.webdriver.common.by import By
import time

# first connect to the driver-
driver=webdriver.Chrome(r"D:\chromedriver_win32.zip\chromedriver.exe")
driver.maximize_window()
#opening the page on automated chrome browser
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]

#scraping rank from the given page
rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for r in rank[0:30]:
    ranks=r.text
    Rank.append(ranks)
#scraping name from the given page
name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for n in name[0:30]:
    names=n.text
    Name.append(names)
#scraping artist from the given page
artist=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for a in artist[0:30]:
    art=a.text
    Artist.append(art)
#scraping uploaddate from the given page
uploaddate=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for u in uploaddate[0:30]:
    upload=u.text
    Upload_date.append(upload)
#scraping views from the given page 
views=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for v in views[0:30]:
    view=v.text
    Views.append(view)
```

```python
#make a dataframe from the above data
df=pd.DataFrame({'Rank':Rank[0:30],'Name':Name[0:30],'Artist':Artist[0:30],'Upload_date':Upload_date,'Views':Views})
df
```
